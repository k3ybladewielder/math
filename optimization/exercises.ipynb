Etapas para resolução:
1. definir variáveis de decisão; 
2. definir o objetivo ou meta em termos das variáveis de decisão; 
3. definir todas as restrições do sistema, e;
4. restringir todas as variáveis para que não sejam negativas. 

## Bonuses and Merits

**Variáveis de decisão:**
- $ X_1 $: valor em dólares de 1 ponto da variável CM1  
- $ X_2 $: valor em dólares de 1 ponto da variável CM2  
- $ X_3 $: valor em dólares de 1 ponto da variável CM3  

**Função objetivo:**
O objetivo é **maximizar o valor de um dos pontos (idealmente de todos, respeitando as proporções)**, ou seja:

$$ \text{Maximizar } Z = X_1 + X_2 + X_3 $$

**Restrições:**

1. **Regra 1 – Nenhum professor pode receber mais de \$5.000:**

$$
5X_1 + 5X_2 + 5X_3 \leq 5000
\quad \text{(limite máximo por professor)}
$$

2. **Orçamento total de \$1.500.000:**

$$
1625X_1 + 1409X_2 + 1387X_3 \leq 1500000
\quad \text{(limite do orçamento total)}
$$

3. **Regra 2 – Proporção entre os valores dos pontos:**
- $ X_1 = 2X_2 $
- $ X_1 = 2X_3 $


**Restrições de não negatividade:**
$$
X_1, X_2, X_3 \geq 0
$$

---

**Resolvendo o problema**

Como temos as equações de proporcionalidade, podemos **substituir $ X_2 $ e $ X_3 $ em função de $ X_1 $**.

Sabemos que:
- $ X_2 = \frac{1}{2}X_1 $
- $ X_3 = \frac{1}{2}X_1 $

**Substituindo na função objetivo:**
$$
Z = X_1 + \frac{1}{2}X_1 + \frac{1}{2}X_1 = 2X_1
\Rightarrow \text{Maximizar } Z = 2X_1
$$

**Substituindo nas restrições:**

1. Limite por professor:
$$
5X_1 + 5\left(\frac{1}{2}X_1\right) + 5\left(\frac{1}{2}X_1\right) \leq 5000
\Rightarrow 5X_1 + 2.5X_1 + 2.5X_1 = 10X_1 \leq 5000
\Rightarrow X_1 \leq 500
$$

2. Orçamento:
$$
1625X_1 + 1409\left(\frac{1}{2}X_1\right) + 1387\left(\frac{1}{2}X_1\right) \leq 1500000
\Rightarrow 1625X_1 + 704.5X_1 + 693.5X_1 = 3023X_1 \leq 1500000
\Rightarrow X_1 \leq \frac{1500000}{3023} \approx 496.26
$$

**Conclusão:**
- A menor entre as duas restrições é o valor que limita:  
  $ X_1 \leq 496.26 $
- Como queremos **maximizar $ X_1 $**, o melhor valor possível é:  
  $ X_1 = 496.26 $
- Então:
  - $ X_2 = \frac{1}{2} \cdot 496.26 = 248.13 $
  - $ X_3 = 248.13 $
  - $ Z = 2X_1 = 992.52 $

---

**Interpretação dos resultados**

- **Variáveis de decisão:**
  - $ X_1 = 496.26 $: Cada ponto CM1 vale \$496,26.
  - $ X_2 = X_3 = 248.13 $: Cada ponto CM2 e CM3 vale \$248,13.

- **Função objetivo (valor total dos pontos):**
  - $ Z = X_1 + X_2 + X_3 = 992.52 $

- **Restrições ativas:**
  - O orçamento total de \$1.500.000 foi completamente usado.
  - A restrição de \$5.000 por professor **não foi atingida**, pois $ 10X_1 = 4962.6 < 5000 $. Portanto, há **folga** nessa restrição.

- **Surplus/excess (folgas):**
  - **Restrição 1 (professor):** folga = $ 5000 - 10X_1 = 37.4 $
  - **Restrição 2 (orçamento):** sem folga, está **saturada**.

```python
from scipy.optimize import linprog

# Como X2 = 0.5*X1 e X3 = 0.5*X1, podemos reescrever o problema com uma variável: X1

# Função objetivo (Maximizar Z = X1 + X2 + X3 = 2X1)
# Como linprog faz minimização, usamos -2*X1 para maximizar
c = [-2]

# Restrições:
# 1) 10 * X1 <= 5000 (Regra 1 reescrita)
# 2) 3023 * X1 <= 1500000 (Restrição do orçamento)
A = [[10],
     [3023]]
b = [5000, 1500000]

# Limite inferior: X1 >= 0
bounds = [(0, None)]

# Resolver o problema
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# Recuperar os valores
X1 = result.x[0]
X2 = 0.5 * X1
X3 = 0.5 * X1
Z = X1 + X2 + X3  # Ou 2*X1

X1, X2, X3, Z, result.success
```

Resolução direta

```python
from scipy.optimize import linprog

# ------------------------------------------
# DEFINIÇÃO DO PROBLEMA
# ------------------------------------------

# Função objetivo:
# Queremos maximizar Z = X1 + X2 + X3
# Como linprog faz minimização, usamos os coeficientes negativos: -Z = -X1 - X2 - X3
c = [-1, -1, -1]

# ------------------------------------------
# RESTRIÇÕES DE DESIGUALDADE (<=)
# ------------------------------------------

# Regra 1: Cada professor pode receber no máximo $5000
# Como são 5 pontos no máximo (de cada tipo), temos: 5X1 + 5X2 + 5X3 <= 5000
# Orçamento total: 1625X1 + 1409X2 + 1387X3 <= 1500000
A = [
    [5, 5, 5],               # Restrição por professor
    [1625, 1409, 1387]       # Restrição do orçamento total
]
b = [5000, 1500000]         # Lados direitos das desigualdades

# ------------------------------------------
# RESTRIÇÕES DE IGUALDADE (==)
# ------------------------------------------

# Regra 2: Proporção entre os pontos
# X1 = 2X2  →  X1 - 2X2 = 0
# X1 = 2X3  →  X1 - 2X3 = 0
A_eq = [
    [1, -2, 0],  # X1 - 2X2 = 0
    [1, 0, -2]   # X1 - 2X3 = 0
]
b_eq = [0, 0]

# ------------------------------------------
# LIMITES DAS VARIÁVEIS
# ------------------------------------------

# X1, X2 e X3 devem ser >= 0 (não faz sentido valores negativos)
bounds = [(0, None), (0, None), (0, None)]

# ------------------------------------------
# RESOLUÇÃO DO PROBLEMA
# ------------------------------------------

# Executa a otimização com o método 'highs' (mais moderno e eficiente)
result_full = linprog(c, A_ub=A, b_ub=b, A_eq=A_eq, b_eq=b_eq, bounds=bounds, method='highs')

# ------------------------------------------
# EXTRAÇÃO DOS RESULTADOS
# ------------------------------------------

# Valores ótimos das variáveis de decisão
X1_full = result_full.x[0]  # Valor ideal para 1 ponto CM1
X2_full = result_full.x[1]  # Valor ideal para 1 ponto CM2
X3_full = result_full.x[2]  # Valor ideal para 1 ponto CM3

# Valor total da função objetivo (soma dos valores dos 3 tipos de pontos)
Z_full = X1_full + X2_full + X3_full

# Exibe os resultados
X1_full, X2_full, X3_full, Z_full, result_full.success
```
